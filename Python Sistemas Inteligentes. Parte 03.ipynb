```python
def f0(x):
    "Compute x^2 - 1"
    return x*x - 1


def f1(x):
    "Compute -x^2 + 2x + 1"
    return -x*x + 2*x + 1


def is_positive(f, x):
    "Check if the function value f(x) is positive"

    # Evaluate the function passed into the function for the value of x 
    # passed into the function
    if f(x) > 0:
        return True
    else:
        return False

    
# Value of x for which we want to test a function sign
x = 4.5

# Test function f0
print(is_positive(f0, x))

# Test function f1
print(is_positive(f1, x))
```

```python
def compute_max_min_mean(x0, x1, x2):
    "Return maximum, minimum and mean values, devoluci√≥n de varios valores"
    
    x_min = x0
    if x1 < x_min:
        x_min = x1
    if x2 < x_min:
        x_min = x2

    x_max = x0
    if x1 > x_max:
        x_max = x1
    if x2 > x_max:
        x_max = x2

    x_mean = (x0 + x1 + x2)/3    
        
    return x_min, x_max, x_mean


xmin, xmax, xmean = compute_max_min_mean(0.5, 0.1, -20)
print(xmin, xmax, xmean)
```

```python
# funci√≥n no recursiva. Se calcula con un bucle for..
def fib(n):
    "Compute the nth Fibonacci number"
    # Starting values for f0 and f1
    f0, f1 = 0, 1

    # Handle cases n==0 and n==1
    if n == 0:
        return 0
    elif n == 1:
        return 1
    
    # Start loop (from n = 2)    
    for i in range(2, n + 1):
        # Compute next term in sequence
        f = f1 + f0

        # Update f0 and f1    
        f0 = f1
        f1 = f

    # Return Fibonacci number
    return f

print(fib(10))
```

Dado que la secuencia de Fibonacci tiene una estructura recursiva, con el ùëõ¬∫ t√©rmino calculado a partir de los t√©rminos ùëõ-1 y ùëõ-2, podr√≠amos escribir una funci√≥n que utilice esta estructura recursiva:

```python
# funci√≥n no recursiva.
def f(n): 
    "Compute the nth Fibonacci number using recursion"
    if n == 0:
        return 0  # This doesn't call f, so it breaks out of the recursion loop
    elif n == 1:
        return 1  # This doesn't call f, so it breaks out of the recursion loop
    else:
        return f(n - 1) + f(n - 2)  # This calls f for n-1 and n-2 (recursion), and returns the sum 

print(f(10))
```

Para utilizar una funci√≥n de un m√≥dulo tenemos que hacerla disponible en nuestro programa.

Al prefijar `sqrt` con `math`, estamos utilizando un *espacio de nombres (namespace)* (que en este caso es `math`).
Esto deja claro qu√© funci√≥n `sqrt` queremos usar - podr√≠a haber m√°s de una funci√≥n `sqrt` disponible.

>*Namespace:* El prefijo `math` indica qu√© funci√≥n `sqrt` queremos utilizar. Este
puede parecer pedante, pero en la pr√°ctica a menudo hay diferentes algoritmos para realizar la misma o similar operaci√≥n. Pueden variar en velocidad y precisi√≥n. En algunas aplicaciones podemos necesitar un m√©todo preciso (pero lento) para calcular la ra√≠z cuadrada, mientras que para otras aplicaciones podemos necesitar velocidad con un compromiso en la precisi√≥n. Pero, si dos funciones tienen el mismo nombre y no se distinguen por un espacio de nombres, tenemos un *choque de nombres*.

> En un programa grande, dos desarrolladores pueden elegir el mismo nombre para dos funciones que realizan tareas similares pero ligeramente diferentes. Si estas funciones est√°n en diferentes m√≥dulos, no habr√° un choque de nombres ya que el nombre del m√≥dulo proporciona un "espacio de nombres" - un prefijo que proporciona una distinci√≥n entre las dos funciones. Los espacios de nombres son muy √∫tiles para los programas de varios autores. Uno de los puntos d√©biles de los lenguajes m√°s antiguos, como C y Fortran, es que no admiten espacios de nombres.
```python
from math import sqrt as some_math_function 

x = 2.0
x = some_math_function(x)
print(x)
```

Renombrar las funciones en la importaci√≥n puede ser √∫til para mantener el c√≥digo corto, y veremos m√°s adelante que puede ser √∫til para cambiar entre diferentes funciones.
Sin embargo, la elecci√≥n del nombre de arriba es una pr√°ctica muy pobre - el nombre 'funci√≥n_matem√°tica` no es descriptivo.
A continuaci√≥n se muestra un ejemplo m√°s sensato.

```python
from math import sqrt as square_root

def compute_roots(a, b, c):
    "Compute roots of the polynomial f(x) = ax^2 + bx + c"
    
    root0 = (-b + square_root(b*b - 4*a*c))/(2*a)
    root1 = (-b - square_root(b*b - 4*a*c))/(2*a)
    
    return root0, root1

# Compute roots of f = 4x^2 + 10x + 1
root0, root1 = compute_roots(4, 10, 1)
print(root0, root1)
```

### Ejemplo de rendimiento: c√°lculo de la norma de un vector largo

La norma de un vector $x$ viene dada por: 

$$
\| x \| = \sqrt{\sum_{i=0}^{n-1} x_{i} x_{i}}
$$

donde $x_{i}$ es la entrada $i$ de $x$. Es el producto punto de un vector consigo mismo, 
seguido de la ra√≠z cuadrada.
Para calcular la norma, podr√≠amos hacer un bucle/iteraci√≥n sobre las entradas del vector y sumar el cuadrado de cada entrada, y luego tomar la ra√≠z cuadrada del resultado.

Vamos a evaluar la norma utilizando dos m√©todos para calcular la norma de una matriz de 10 millones de longitud para comparar su rendimiento. Primero creamos un vector con 10 millones de entradas aleatorias, utilizando NumPy:



```python
# Create a NumPy array with 10 million random values
x = np.random.rand(10000000)
print(type(x))
```

Ahora cronometramos el tiempo que se tarda en calcular la norma del vector utilizando la funci√≥n NumPy '`numpy.dot`'. Usamos el "comando m√°gico" de Jupyter [`%time`](Notebook%20tips.ipynb#Simple-timing) para cronometrar la operaci√≥n: 

```python
%time norm = np.sqrt(np.dot(x, x))
print(norm)
```

La salida de tiempo de inter√©s es `Wall time`.

>  U[texto del enlace](https://)samos `%time` como una herramienta compacta y coveniente para 
> medir el tiempo que tarda en ejecutarse un comando.

Ahora realizamos la misma operaci√≥n con nuestra propia funci√≥n para calcular la norma:

```python
def compute_norm(x):
    norm = 0.0
    for xi in x:
        norm += xi*xi
    return np.sqrt(norm)

%time norm =compute_norm(x)
print(norm)
```

Deber√≠as ver que los dos enfoques dan el mismo resultado, pero la funci√≥n 
NumPy es m√°s de 100 veces m√°s r√°pida, ¬°y posiblemente m√°s de 100.000 veces m√°s r√°pida!

El mensaje es que las funciones y estructuras de datos especializadas para c√°lculos num√©ricos pueden ser muchos √≥rdenes de magnitud m√°s r√°pidas que sus propias implementaciones generales. Adem√°s, las funciones especializadas son mucho menos 
de tener errores.

### Operaciones con matrices en 2D

Para aquellos que hayan visto matrices anteriormente, las operaciones de esta secci√≥n les resultar√°n familiares. Para aquellos que no han visto las matrices, es posible que desee volver a esta secci√≥n una vez que las matrices se han cubierto en las clases de matem√°ticas.

#### Multiplicaci√≥n matriz-vector y matriz-matriz

Consideraremos la matriz $A$:
$$
A  = 
\begin{bmatrix}
3.4 & 2.6 \\
2.1 & 4.5
\end{bmatrix}
$$

and the vector $x$:

$$
x  = 
\begin{bmatrix}
0.2 \\ -1.1
\end{bmatrix}
$$

```python
A = np.array([[3.4, 2.6], [2.1, 4.5]])
print("Matrix A:\n {}".format(A))

x = np.array([0.2, -1.1])
print("Vector x:\n {}".format(x))
```

Podemos calcular el producto matriz-vector $y = Ax$ mediante:

# Extracci√≥n de matrices (Array Slicing)

Cuando se trabaja con matrices, a menudo es √∫til extraer subconjuntos de una matriz. Podemos querer s√≥lo las 3 primeras entradas de un array largo, o podemos querer la segunda columna de un array 2D (matriz). Estas operaciones se conocen como *array slicing* (https://en.wikipedia.org/wiki/Array_slicing).

Exploraremos el slicing a trav√©s de ejemplos. Comenzamos creando un array lleno de valores aleatorios:

```python
x = np.random.rand(5)
print(x)
```

A continuaci√≥n se muestran algunas operaciones de corte:

```python
# Using ':' implies the whole range of indices, i.e. from 0 -> (length-1)
y = x[:]
print("Slice using '[:]' {}".format(y))

# Using '1:3' implies indices 1 -> 3 (not including 3)
y = x[1:3]
print("Slice using '[1:3]': {}".format(y))

# Using '2:-1' implies indices 2 -> second-from-last
y = x[2:-1]
print("Slice using '[2:-1]': {}".format(y))

# Using '2:-2' implies indices 2 -> third-from-last
y = x[2:-2]
print("Slice using '[2:-2]': {}".format(y))
```

> N√≥tese el uso del √≠ndice `-1`. El √≠ndice `-1` corresponde a la √∫ltima entrada de la matriz, y `-2` a la 
> la pen√∫ltima entrada, etc. Esto es conveniente si sabemos a qu√© distancia del final de un array se encuentra una entrada deseada.
> Usando √≠ndices negativos podemos expresar esto sin referencia a la longitud del array.

Si queremos que un slice vaya desde el principio de un array, o hasta el final de un array, lo hacemos: 

```python
# Using ':3' implies start -> 3 (not including 3)
y = x[:3]
print("Slice using '[:3]': {}".format(y))

# Using '4:' implies 4 -> end
y = x[4:]
print("Slice using '[4:]': {}".format(y))

# Using ':' implies start -> end
y = x[:]
print("Slice using '[:]': {}".format(y))
```

El corte puede aplicarse a matrices 2D:

```python
B = np.array([[1.3, 0], [0, 2.0]])
print(B)

# Extract second row
r = B[1, :]
print(r)
```

# Ejemplo de m√≥dulo: procesamiento en paralelo

Los m√≥dulos est√°ndar pueden simplificar problemas muy t√©cnicos. Un ejemplo es el procesamiento paralelo.

La mayor√≠a de las CPUs -desde los tel√©fonos hasta los superordenadores- tienen ahora CPUs con m√∫ltiples n√∫cleos, y cada n√∫cleo realiza c√°lculos. Para beneficiarse de los m√∫ltiples n√∫cleos, tenemos que calcular en *paralelo*.
Un programa "est√°ndar" realiza las tareas en orden, y en este caso s√≥lo se utilizar√° un n√∫cleo y el resto permanecer√°
permanecer√° inactivo. Para obtener el mejor rendimiento del hardware, tenemos que calcular en paralelo. Es decir, realizar varias tareas al mismo tiempo. 

El procesamiento paralelo es un tema enorme en s√≠ mismo, pero podemos tocarlo aqu√≠ porque tenemos bibliotecas est√°ndar
que facilitan su uso. La gesti√≥n de tareas paralelas a bajo nivel es extremadamente t√©cnica, pero las bibliotecas est√°ndar pueden facilitarla. Utilizaremos el m√≥dulo `multiprocessing`, y lo usaremos para ordenar listas de n√∫meros de forma concurrente.

Empezaremos por ver c√≥mo generar una lista de enteros aleatorios utilizando el m√≥dulo `random`. El siguiente c√≥digo 
crea una lista (m√°s sobre listas en el siguiente cuaderno) de 10 enteros aleatorios en el rango de 0 a 100 (sin incluir el 100):

```python
import random
x = random.sample(range(0, 100), 10)
print(x)
```

Para crear una lista ordenada, utilizamos la funci√≥n incorporada `sorted`:

```python
y = sorted(x)
print(y)
```

Ahora, si necesitamos ordenar varias listas diferentes, podr√≠amos ordenar las listas una tras otra, o podr√≠amos ordenar varias listas al mismo tiempo (en paralelo). Nuestro sistema operativo gestionar√° entonces el env√≠o de la tarea de ordenaci√≥n a los diferentes n√∫cleos del procesador. Antes de ver c√≥mo hacer esto, implementamos una funci√≥n para realizar la ordenaci√≥n:

```python
import multiprocessing
import random

def mysort(N):
    "Create a list of random numbers of length N, and return a sorted list"

    # Create random list 
    x = random.sample(range(0, N), N) 

    # Print process identifier (just out of interest)
    print("Process id: {}".format(multiprocessing.current_process()))
    
    # Return sorted list of numbers
    return sorted(x)
```

Para crear las listas ordenadas, poniendo a disposici√≥n tres procesos (hilos), utilizamos:

```python
N = 20000
with multiprocessing.Pool(processes=3) as p:
    p.map(mysort, [N, N, N])  # Call function mysort three times
```

Vemos en la salida que tres procesos diferentes han trabajado en nuestro problema - uno para cada tarea de clasificaci√≥n.

Utilizamos el procesamiento en paralelo para hacer los c√°lculos m√°s r√°pidos. Vamos a cronometrar nuestro c√°lculo utilizando diferentes n√∫meros de procesos para ver c√≥mo afecta al rendimiento. Para realizar el cronometraje, primero encapsulamos nuestro problema en una funci√≥n:

```python
def parallel_sort(N, num_proc):
    "Create three lists of random numbers (each of length N) using num_proc processes"
    with multiprocessing.Pool(processes=num_proc) as p:
        p.map(mysort, [N, N, N])
```

Usando el comando '%time', cronometramos la ordenaci√≥n usando s√≥lo un proceso (el √∫nico proceso ordena las listas una tras otra):

```python
N = 500000
%time parallel_sort(N, 1)    
```

Vemos por el `Process id` que tres procesos diferentes han trabajado en las listas. La ejecuci√≥n en paralelo deber√≠a ser m√°s r√°pida (esto depender√° del hardware).
